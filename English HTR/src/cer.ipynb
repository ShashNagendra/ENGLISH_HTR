```python
import os
import re
import statistics

def extract_metrics(file_path):
    with open(file_path, 'r') as f:
        content = f.read()

    cer_match = re.search(r'Character Error Rate:\s+(\d+\.\d+)', content)
    wer_match = re.search(r'Word Error Rate:\s+(\d+\.\d+)', content)
    ser_match = re.search(r'Sequence Error Rate:\s+(\d+\.\d+)', content)

    if cer_match and wer_match and ser_match:
        cer = float(cer_match.group(1))
        wer = float(wer_match.group(1))
        ser = float(ser_match.group(1))
        return cer, wer, ser
    else:
        return None

def process_folders(root_folder):
    cer_values = []
    wer_values = []
    ser_values = []
    evaluate_count = 0
    
    for root, dirs, files in os.walk(root_folder):
        if 'evaluate.txt' in files:
            evaluate_count += 1
            file_path = os.path.join(root, 'evaluate.txt')
            metrics = extract_metrics(file_path)
            if metrics is not None:
                cer, wer, ser = metrics
                cer_values.append(cer)
                wer_values.append(wer)
                ser_values.append(ser)

                # Print values for each evaluate.txt file
                print(f"File: {file_path}")
                print(f"Character Error Rate (CER): {cer:.6f}")
                print(f"Word Error Rate (WER): {wer:.6f}")
                print(f"Sequence Error Rate (SER): {ser:.6f}")
                print("=" * 20)
    
    return cer_values, wer_values, ser_values, evaluate_count

root_folder = '/home/project/Desktop/shashank/English HTR/output/bentham/shashankbest4senetgateddropouttwo'
cer_values, wer_values, ser_values, evaluate_count = process_folders(root_folder)

cer_avg = statistics.mean(cer_values)
cer_stddev = statistics.stdev(cer_values)

wer_avg = statistics.mean(wer_values)
wer_stddev = statistics.stdev(wer_values)
 
ser_avg = statistics.mean(ser_values)
ser_stddev = statistics.stdev(ser_values)

print(f"Total number of evaluate.txt files: {evaluate_count}")
print(f"Average Character Error Rate (CER): {cer_avg:.6f}")
print(f"Standard Deviation of CER: {cer_stddev:.6f}")
print(f"Average Word Error Rate (WER): {wer_avg:.6f}")
print(f"Standard Deviation of WER: {wer_stddev:.6f}")
print(f"Average Sequence Error Rate (SER): {ser_avg:.6f}")
print(f"Standard Deviation of SER: {ser_stddev:.6f}")



print(cer_values)
```
